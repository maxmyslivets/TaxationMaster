```python
from IPython.display import display

from pathlib import Path

import pandas as pd

from src.processing.taxation_list import _open_file_taxation_list
```

```python
file_path = Path().absolute().parent / "data/example_1_taxation_list.xlsx"

is_import_first_row = False
column_mapping = {
    'number': 0,
    'name': 1,
    'quantity': 2,
    'height': 3,
    'diameter': 4,
    'quality': 5
}

taxation_data_df = _open_file_taxation_list(file_path, is_import_first_row, column_mapping)
taxation_data_df
```

```python
tree_size_df = taxation_data_df[['origin_number', 'name', 'quantity', 'height', 'diameter']].copy(deep=True)
tree_size_df
```

```python
def split_tree_size(origin_number: str, name: str, quantity: str, height: str, diameter: str) -> list[list[str]]:
    return [[origin_number, origin_number, name, quantity, height, diameter]]
```

```python
processed_data = []
for _, row in tree_size_df.iterrows():
    split_tree_size_result = split_tree_size(row['origin_number'], row['name'], row['quantity'], row['height'], row['diameter'])
    for origin_number, splitted_number, name, quantity, height, diameter in split_tree_size_result:
        processed_data.append(
            {
                'origin_number': origin_number,
                'split_number': splitted_number,
                'name': name,
                'quantity': quantity,
                'height': height,
                'diameter': diameter
            }
        )
processed_data_df = pd.DataFrame(processed_data)
processed_data_df
```
